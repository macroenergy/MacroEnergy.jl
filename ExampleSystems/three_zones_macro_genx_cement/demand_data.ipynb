```python
import pandas as pd
import numpy as np
```

```python
df_monthly = pd.read_csv("system/monthly_cement_demand.csv")
df_monthly.drop("Time_Index", axis=1, inplace=True)
dates = pd.date_range('2021-01-01', periods=12, freq='MS')
df_monthly.index = dates
```

```python
df_monthly
```

```python
days_per_month = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])
df_hourly= df_monthly.div(days_per_month * 24, axis=0)
df_hourly
```

```python
df_hourly = df_hourly.asfreq(freq='H')
df_hourly
```

```python
missing_dates = pd.date_range(start="2021-12-01 01:00:00", end="2021-12-31 23:00:00", freq="H")
missing_df = pd.DataFrame({"Cement_Demand_Tons_MA": [None] * len(missing_dates), 
                          "Cement_Demand_Tons_CT": [None] * len(missing_dates),
                          "Cement_Demand_Tons_ME": [None] * len(missing_dates)}, index=missing_dates)
df_hourly = pd.concat([df_hourly, missing_df], ignore_index=True)
```

```python
df_hourly.index = pd.date_range(start="2021-01-01 00:00:00", end="2021-12-31 23:00:00", freq="H")
```

```python
df_hourly = df_hourly.ffill()
df_hourly
```

```python
df_hourly.groupby(df_hourly.index.month).sum()
```

```python
df_hourly.insert(0, "Time_Index", np.arange(1, 8761, 1))
```
